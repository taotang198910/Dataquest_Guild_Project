# 1. Cleaning Data

## Read in the data

** Reading each file into a pandas dataframe, and then store all of the dataframes in a dictionary. This will give us a conveninent way to store them, and a quick way to reference them later on. **

```python
import pandas as pd
import numpy as np
import re
import glob
import matplotlib.pyplot as plt
%matplotlib inline

data_files=glob.glob("schools/*.csv")
data = {}

for f in data_files:
    d = pd.read_csv(f)
    key_name=f.replace('.csv','')
    key_name=key_name.replace('schools/','')
    data[key_name] = d

data_files
data['sat_results'].head()
```

```python
for i in data:
    print(data[i].head())
```

____
** Each data set appears to either have a DBN column, or the information we need to create one. That means we can use a DBN column to combine the data sets.**
_____

## Read in the surveys

```python
all_survey = pd.read_csv("schools/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("schools/survey_d75.txt", delimiter="\t", encoding='windows-1252')
survey = pd.concat([all_survey, d75_survey], axis=0)

##Create a new column DBN
survey["DBN"] = survey["dbn"]

## Filter the data to remove the unnecessary columns we don't need.
survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]
survey = survey.loc[:,survey_fields]
data["survey"] = survey
data['survey'].head()
```

_____
## Add DBN columns

** When we explored all the data sets, we noticed that some of them, like <font color='red'>class_size, hs_directory</font>  don't have a DBN column. **

```python
data['hs_directory'].head(2)
```

```python
data['hs_directory']['DBN']=data['hs_directory']['dbn']
```

```python
data['class_size'].head(2)
```

** From looking at these rows, we can tell that the <font color='red'>DBN</font> is just a conbination of <font color='red'>CSD</font> and <font color='red'>SCHOOL CODE</font>. The main difference is that the DBN is padded, so that the CSD portion of it always consists of two digits. **

```python
def pad_csd(num):
    string=str(num)
    if len(string)>1:
        return string
    else:
        return string.zfill(2)

data['class_size']['DBN']=data['class_size']['CSD'].apply(pad_csd)+data['class_size']['SCHOOL CODE']
print(data['class_size']['DBN'].head())
    
```

____
## Convert columns to numeric

```python
data['sat_results'].info()
```

```python
##Converting object to numeric.
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")

#Create a new column sat_score
data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]
```

```python
## Parsing the latitude and longitude coordinates for each school. This will enable us to map the schools and uncover any geographic patterns in the data.
data['hs_directory']['Location 1'].head()
```

```python
import re 
def find_lat(loc):
    coords=re.findall('\(.+\)',loc)
    lat=coords[0].split(',')[0].replace('(','')
    return lat

def find_lon(loc):
    coords=re.findall('\(.+\)',loc)
    lon=coords[0].split(',')[1].replace(')','')
    return lon

data['hs_directory']['lat']=data['hs_directory']['Location 1'].apply(find_lat)
data['hs_directory']['lon']=data['hs_directory']['Location 1'].apply(find_lon)
data['hs_directory']['lat']=pd.to_numeric(data['hs_directory']['lat'],errors='coerce')
data['hs_directory']['lon']=pd.to_numeric(data['hs_directory']['lon'],errors='coerce')


data['hs_directory'][['lat','lon']].info()
```

____
## Condense datasets

** Noticed that the values in the DBN column were unique in the <font color='red'>sat_results</font>. Other data sets like <font color='red'>class_size</font> had duplicate DBN values.
To resolve this issue, we condensing the <font color='red'>class_size, graduation, demographics</font>.**

```python
data['class_size'].head()
```

** As you can see, the first few rows all pertain to the same school, which is why the DBN appears more than once. It looks like each school has multiple values for <font color='red'>GRADE, PROGRAM TYPE, CORE SUBJECT (MS CORE and 9-12 ONLY), and CORE COURSE (MS CORE and 9-12 ONLY)</font>. Because we're dealing with high schools, so only concerned with grades 9 through 12. Each school have multiple program types, let's only focued on <font color='red'>GEN ED</font>. **

```python
## Unique vaues for GRADE and PROGRAM TYPE
data['class_size']['GRADE '].value_counts()
data['class_size']['PROGRAM TYPE'].value_counts()
```

```python
class_size=data['class_size']
class_size=class_size[(class_size['GRADE ']=='09-12') & (class_size['PROGRAM TYPE']=='GEN ED')]

class_size=class_size.groupby('DBN').agg(np.mean)
class_size.head(5)
```

```python
class_size.reset_index(inplace=True)
data['class_size']=class_size
data['class_size'].head(5)
```

```python
## Select rows where shcoolyear is 20112012
data['demographics']=data['demographics'][data['demographics']['schoolyear']==20112012]
```

```python
data['demographics'].head()
```

```python
data['graduation'].head()
```

```python
data['graduation']['Demographic'].value_counts()
```

```python
data['graduation']=data['graduation'][(data['graduation']['Cohort']=='2006') & (data['graduation']['Demographic']=='Total Cohort')]
```

____
## Convert AP scores to numeric

```python
data['ap_2010'].info()
```

```python
cols=['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']
for col in cols:
    data['ap_2010'][col]=pd.to_numeric(data['ap_2010'][col],errors='coerce')
```

_____
## Combine the datasets

```python
## Left join sat_results, ap_2010 and graduation
combined=data['sat_results'].merge(data['ap_2010'],how='left',on='DBN')
combined=combined.merge(data['graduation'],how='left',on='DBN')
combined.shape

##Inner join class_size, demographics, surver, and hs_directory
columns=['class_size','demographics','survey','hs_directory']
for col in columns:
    combined=combined.merge(data[col],how='inner',on='DBN')

combined.shape
```

```python
## As we can see that lot of missing values
import missingno
missingno.matrix(combined)
```

```python
##Filling in the missing values with the overall mean for the column
combined=combined.fillna(combined.mean()).fillna(0)
```

```python
## Adding a column to the data stet that specifies the school district that will help us to map the statistics out on a school district
## School district is just the first two characters of the DBN.
def school(str):
    return str[:2]

combined['school_district']=combined['DBN'].apply(school)
combined['school_district'].value_counts()
```
