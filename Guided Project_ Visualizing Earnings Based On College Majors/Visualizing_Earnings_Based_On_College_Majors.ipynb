# Visualizing Earnings Based On College Majors

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads=pd.read_csv('recent-grads.csv')
```

```python
recent_grads=recent_grads.dropna()
```

```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter')
```

```python
recent_grads.plot('Sample_size','Unemployment_rate',kind='scatter')
```

```python
recent_grads.plot('Full_time','Median',kind='scatter')
```

```python
recent_grads.plot('ShareWomen','Unemployment_rate',kind='scatter')
```

```python
recent_grads.plot('Men','Median',kind='scatter')
```

```python
recent_grads.plot('Women','Median',kind='scatter')
```

```python
columns=['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig=plt.figure(figsize=(6,24))
for r in range(1,9):
    ax=fig.add_subplot(8,2,r)
    ax=recent_grads[columns[r-1]].plot(kind='hist',rot=40,title=columns[r-1])
    plt.tight_layout()
```

```python
from pandas.tools.plotting import scatter_matrix 
```

```python
scatter_matrix(recent_grads[['Sample_size','Median']])
plt.tight_layout()
```

```python
scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']])
```

```python
recent_grads['ShareWomen'].sort_index().head(10).plot(kind='bar')
plt.xticks(range(0,10),recent_grads['Major'])
```
